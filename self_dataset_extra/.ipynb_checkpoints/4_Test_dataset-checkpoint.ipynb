## Dataset

```python
model_ds=model.ModelDatasetComponent(cfg, True)

tfrecord_files_train=dataset_builder_self.get_tfrecord_filenames(cfg.TRAIN_TFRECORDS,cfg)
print("Found the following training TFrecords:\n","\n".join(tfrecord_files_train))

print("Building training dataset")
dst=dataset_builder_self.build_training_ds(tfrecord_files_train,model_ds.place_training_labels,cfg)
print("Training dataset shape:",dst)
```

```python
#Test element
print("Dataset shape:",dst) #this should match the model input, and output stages
dst_iter=iter(dst)
```

```python
pic_idx = 0
test_type = 'paf'
keypoint_dix = 15
sample_elem=next(dst_iter)

#testing img
img = sample_elem[0][0][pic_idx,...,:3]

#testing mask
if test_type == 'kpt':
    m=sample_elem[1][4][pic_idx,...,keypoint_dix]
    m_big = cv2.resize(m.numpy(), dsize = (765, 368))
elif test_type == 'paf':
    m=sample_elem[1][3][pic_idx,...,keypoint_dix]
    m_big = cv2.resize(m.numpy(), dsize = (765, 368))

plt.figure(figsize=[15,15])
plt.imshow(img)
plt.imshow(m_big, alpha = 0.6)
```

```python
img_numpy = np.round(img.numpy() * 255).astype('uint8')
kpts = sample_elem[1][5][pic_idx,...,:-1].numpy()
pafs = sample_elem[1][3][pic_idx,...,:-1].numpy()
```

```python
skeletonizer = post.Skeletonizer(kpts, pafs)
skeletons = skeletonizer.create_skeletons()
skeleton_drawer = vis.SkeletonDrawer(img_numpy, draw_config)
for skeleton in skeletons:
    skeleton.draw_skeleton(skeleton_drawer.joint_draw, skeleton_drawer.kpt_draw)
```

```python
plt.figure(figsize=[15,15])
plt.imshow(img_numpy)
```
